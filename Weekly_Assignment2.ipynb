```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical

file_path = '/content/drive/MyDrive/Workspace/drug200.csv'
df = pd.read_csv(file_path)

X = df.drop('Drug', axis=1)
y = df['Drug']


le = LabelEncoder()
X['Sex'] = le.fit_transform(X['Sex'])

X = pd.get_dummies(X, columns=['BP', 'Cholesterol'])


le_y = LabelEncoder()
y = le_y.fit_transform(y)
y_one_hot = to_categorical(y)


X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.2, random_state=42)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(len(le_y.classes_), activation='softmax'))


model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy * 1:.2f}')

```

```python
import pandas as pd
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
Vectorizer = TfidfVectorizer()

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')


sms = pd.read_csv('/content/drive/MyDrive/Workspace/sentiment_tweets3.csv')
sms = sms.drop(['Index'], axis = 1)

def text_process(mess):
    tokens = nltk.word_tokenize(mess)
    stop_words = set(stopwords.words('english')+[':','/','!','#','$','%','&','(',')','?','-','=','.',','])
    filtered_tokens = [word for word in tokens if (word.lower() not in stop_words and
                                                   not word.startswith('http') and
                                                   not word.startswith('//') and
                                                   not '.com' in word and
                                                   not '=' in word and
                                                   not len(word)<=2 and
                                                   not '@' in word)]
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
    if len(lemmatized_words)==0:
      return None
    data = Vectorizer.fit_transform(lemmatized_words)
    return data.toarray()

sms['clean_msg'] = sms['message to examine'].apply(text_process)
```
